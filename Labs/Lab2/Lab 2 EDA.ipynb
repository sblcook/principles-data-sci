```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# the following line allows ipython to display plots
%matplotlib inline  
df = pd.read_csv('C:\Users\Samuel\Desktop\School\Spring 2018\Data Science\Labs\Lab2\cars.csv')  # df is DataFrame object
```

```python
df['Make'] = df['Car'].map(lambda x: x.split(' ')[0].strip()) #parse make of car and create a new column in data frame
make_mpg = {} #will hold sum of make MPG, later to divide by amount to calculate the average mpg per make
make_count = {} #to calculate average MPG of make
for car in df.iterrows(): #iterate through df
    if car[1].Make not in make_mpg: 
       make_mpg[car[1].Make] = 0
    if car[1].Make not in make_count:
        make_count[car[1].Make] = 0
    make_mpg[car[1].Make] += car[1].MPG
    make_count[car[1].Make] += 1
    
print(make_mpg)
#make_mpg now holds sum of mpg, make_count holds how many of each make we have seen

for make in make_mpg:
    make_mpg[make] = round(make_mpg[make] / make_count[make], 2)
del make_mpg['Citroen']
```

```python
plt.figure(figsize=(8,8))  # can control the size of the display

lists = sorted(make_mpg.items()) # sorted by key, return a list of tuples

x, y = zip(*lists) # unpack a list of pairs into two tuples

plt.barh(x, y)
plt.title('Manufacturer vs. Average MPG')
plt.xlabel('Average MPG')
plt.savefig('fig1')
```
